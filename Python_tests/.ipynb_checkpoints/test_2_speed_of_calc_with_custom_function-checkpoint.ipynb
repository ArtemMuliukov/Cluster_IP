```python
from sklearn.neighbors import BallTree
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import time
import random
from tqdm import tqdm_notebook

tetta = np.array(range(10,71))
w = np.exp(-2*np.log(tetta/54)**2)/tetta

data = pd.read_csv("borovkova/database_txt/database.txt",delimiter ="\t",header=-1)#ненормированны
data.iloc[:,4:] = data.values[:,4:]*w

data_test = pd.read_csv("borovkova/database_txt/test_1000.txt",delimiter ="\t",header=-1)#нормированны

data_exp = pd.read_csv("borovkova/database_txt/experiment_1000.txt",delimiter ="\t",header=-1)#нормированны
```

```python
plt.plot(data.values[101][4:])
plt.show()
```

```python
def calc_time_for_samples(tree, test_data, deep = 1):
    time.clock()
    summary_time = 0
    start_time = time.clock()
    try:
        dist, ind = tree.query(test_data, k=deep)
    except:
        print("!",end="")
        pass
    summary_time += time.clock() - start_time
    return summary_time
```

```python
class Custom_loss_func:
    def __init__(self):
        self.counter = 0
        
    def __call__(self,A,B):
        self.counter += 1
        return ((A - B) ** 2).mean()
```

```python
ref_points = np.linspace(100000,500000,4).astype(int)
ball_res = []
brute_res = []

ball_comps = []
brute_comps = []

for len_of_data in tqdm_notebook(ref_points):
    ball_loss = Custom_loss_func()
    brute_loss = Custom_loss_func()
    clever_tree = BallTree(data.iloc[:len_of_data,4:], leaf_size=1, metric = ball_loss) 
    stupid_tree = BallTree(data.iloc[:len_of_data,4:], leaf_size=data.shape[0],metric = brute_loss)
    print ("Teached dataset of " + str(len_of_data) + " objects")
    ball_loss.counter = 0
    brute_loss.counter = 0 
    ball_res.append(calc_time_for_samples(clever_tree, data_exp[:10]))
    ball_comps.append(ball_loss.counter)
    print (ball_loss.counter)
    brute_res.append(calc_time_for_samples(stupid_tree, data_exp[:10]))
    brute_comps.append(brute_loss.counter)
    print (brute_loss.counter)
```

```python
plt.plot(ref_points, ball_res)
plt.plot(ref_points, brute_res)
plt.legend(["ball", "brute"])
plt.yscale('log')
plt.show()
```

```python
f, axarr = plt.subplots(1,3, sharex=True,figsize=(20,5))
axarr[0].plot(ref_points, ball_res)
axarr[1].plot(ref_points, brute_res)
axarr[2].plot(ref_points, np.array(brute_res)/np.array(ball_res))
plt.show()
```

```python
f, axarr = plt.subplots(1,3, sharex=True,figsize=(20,5))
axarr[0].plot(ref_points, ball_comps)
axarr[1].plot(ref_points, brute_comps)
axarr[2].plot(ref_points, np.array(brute_comps)/np.array(ball_comps))
plt.show()
```
