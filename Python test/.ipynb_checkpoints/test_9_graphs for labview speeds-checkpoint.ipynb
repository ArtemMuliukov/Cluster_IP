```python
from sklearn.neighbors import BallTree
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import time
import random
from tqdm import tqdm_notebook
import pickle
from  matplotlib import ticker
```

```python
data = pd.read_excel("tests_with_Labview_2.xlsx")
```

```python
data
```

```python
xticks = data.iloc[7,1:7].astype(int).values
speedup_withstat = data.iloc[13,1:7].astype(int).values
speedup_nostat = data.iloc[14,1:7].astype(int).values
```

```python
last_p = 3

z1 = np.polyfit(np.log10(xticks[-last_p:]), np.log10(speedup_withstat[-last_p:]), 1)
p1 = np.poly1d(z1)
print(p1)

z2 = np.polyfit(np.log10(xticks[-last_p:]), np.log10(speedup_nostat[-last_p:]), 1)
p2 = np.poly1d(z2)
print(p2)
```

```python
ax = plt.figure().add_subplot(111)

plt.plot(xticks,speedup_withstat, 'o-')
plt.plot(xticks,speedup_nostat, 'o-')


plt.title("Ускорение в реальной системе")
plt.xscale("log")
plt.xticks([5000, 10000, 20000, 50000, 100000, 200000])
plt.xlabel("Размер базы данных")
plt.yscale("log")
plt.yticks([1, 2 ,3 , 5, 10, 20, 40])
plt.ylabel("Ускорение (во сколько раз)")

points_for_line = np.array([np.log10(xticks[0]), np.log10(xticks[-1])])
plt.plot(10**points_for_line, 10**p1(points_for_line),c ='black')
plt.plot(10**points_for_line, 10**p2(points_for_line),c ='gray')

plt.legend(['С дов. инт.','Без дов. инт.',
            "$k_1\,x^{" + str(p1[1])[:4] + "}$",
            "$k_2\,x^{" + str(p2[1])[:4] + "}$"])


ax.yaxis.set_major_formatter(ticker.ScalarFormatter())
ax.xaxis.set_major_formatter(ticker.ScalarFormatter())

plt.savefig("real speed.svg",format = 'svg')
plt.show()
```
